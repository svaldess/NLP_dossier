# 1) Load data

```python
import os, json
import glob
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
def load_data(path_to_json):
    '''Load all json files and convert them into a DataFrame'''  
    
    df = pd.DataFrame()
    for file in os.listdir(path_to_json):
        input_file = path_to_json + file
        with open(input_file) as f:
            data = pd.read_json(f)
            data['title-searched'] = path_to_json.split('/')[2]
            data['city'] = file.replace('.json','')
            data['position'] = data['position'].str.lower()
            df = df.append(data, ignore_index=True, sort=False)
    return df
```

```python
# Create datascience dataframe
df_original_datascience = load_data('./raw_json_data/data-science/')
df_original_datascience.shape
```

```python
df_original_datascience.head()
```

```python
# Create software engineering dataframe
df_original_engineer = load_data('./raw_json_data/software-engineer/')
df_original_engineer.shape
```

```python
df_original_engineer.head()
```

# 2) Create unique ID's for the documents

```python
ds_key = 'DS'
datascience_id = [ds_key+str(i) for i in range(1,df_original_datascience.shape[0]+1)]
len(datascience_id)
```

```python
df_original_datascience['id'] = datascience_id
#df_original_datascience.head()
```

```python
se_key = 'SE'
engineer_id = [se_key+str(i) for i in range(1,df_original_engineer.shape[0]+1)]
df_original_engineer['id'] = engineer_id
#df_original_engineer.head()
```

# 3) Get positions that match "data science" or "software engineer" 

```python
df_datascience = df_original_datascience[df_original_datascience['position'].str.match('data\sscientist')==True]
```

```python
df_datascience.shape
```

```python
df_engineer = df_original_engineer[df_original_engineer['position'].str.match('software\sengineer')==True]
```

```python
df_engineer.shape
```

# 4) Concat dataframes

```python
df = pd.concat([df_datascience, df_engineer], sort=False)
df.shape
```
